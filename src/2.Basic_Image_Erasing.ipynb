```python
from pathlib import Path 
import torch 
import matplotlib.pyplot as plt 

from torchvision.transforms import v2 
from torchvision.io import read_file, decode_image 

torch.manual_seed(1)

from helpers import * 
img_bytes = read_file('image.png')
orig_img = decode_image(img_bytes)
print(f"{type(orig_img) = }, {orig_img.dtype = }, {orig_img.shape = }")
```

### Cutout and Random Erasing

```python
rotater = v2.RandomErasing(p=0.5, scale=(0.1, 0.33), ratio=(0.3, 3.3), value=0.0)
rotated_imgs = [rotater(orig_img) for _ in range(4)]
plot([orig_img] + rotated_imgs)
```

```python
rotater = v2.RandomErasing(p=0.5, scale=(0.02, 0.33), ratio=(0.3, 0.33), value=0.0)
rotated_imgs = [rotater(orig_img) for _ in range(4)]
plot([orig_img] + rotated_imgs)
```

### Hide-and-Seek

```python
h_imgs = [hide_patch(orig_img, hide_prob=0.5) for _ in range(4)]
plot([orig_img] + h_imgs)
```
